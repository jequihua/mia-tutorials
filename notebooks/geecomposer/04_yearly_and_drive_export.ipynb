# Yearly Composites and Drive Export Tasks

This tutorial shows yearly grouping and Drive export task creation. It does not start exports unless you explicitly opt in.


```python
from pathlib import Path

ROOT = Path.cwd().resolve()
while ROOT.name != "mia-tutorials" and ROOT.parent != ROOT:
    ROOT = ROOT.parent

DATA = ROOT / "data"
REPOS = ROOT.parent
ROOT

from geecomposer import compose_yearly, export_to_drive, initialize

AOI = DATA / "geecomposer" / "sample_aoi.geojson"
YEARS = [2024, 2025]
DRIVE_FOLDER = "mia-tutorials"
START_EXPORTS = False
```

```python
yearly_kwargs = dict(
    years=YEARS,
    dataset="sentinel2",
    aoi=str(AOI),
    mask="s2_cloud_score_plus",
    reducer="median",
)
yearly_kwargs
```

```python
RUN_EE = False

if RUN_EE:
    initialize(project="manglariars", authenticate=False)
    images = compose_yearly(**yearly_kwargs)
    for year, img in images.items():
        task = export_to_drive(
            img,
            description=f"rbmn_s2_median_{year}",
            folder=DRIVE_FOLDER,
            region=str(AOI),
            scale=10,
        )
        if START_EXPORTS:
            task.start()
        print(year, task.status())
else:
    print("Skipped live export task creation. Set RUN_EE = True after credentials are ready.")
```
