# Initialize Earth Engine and Load AOIs

This tutorial introduces the `geecomposer` public API and the local AOI fixtures. It avoids network calls unless you explicitly opt into Earth Engine initialization.


```python
from pathlib import Path

ROOT = Path.cwd().resolve()
while ROOT.name != "mia-tutorials" and ROOT.parent != ROOT:
    ROOT = ROOT.parent

DATA = ROOT / "data"
REPOS = ROOT.parent
ROOT

import json
import geopandas as gpd
from geecomposer import compose, compose_yearly, export_to_drive, initialize

AOI = DATA / "geecomposer" / "sample_aoi.geojson"
MULTI_AOI = DATA / "geecomposer" / "multi_feature_aoi.geojson"
AOI
```

```python
with AOI.open("r", encoding="utf-8") as f:
    geojson = json.load(f)
geojson["type"], geojson.get("features", [geojson])[0]["geometry"]["type"]
```

```python
gdf = gpd.read_file(AOI)
print(gdf.crs)
gdf
```
