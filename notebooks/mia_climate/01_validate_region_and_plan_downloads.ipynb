# Validate Region and Plan ERA5-Land Downloads

This tutorial uses `mia-climate` in dry-run mode: validate a region polygon and build a planned-only ERA5-Land request manifest without touching the network.


```python
from pathlib import Path

ROOT = Path.cwd().resolve()
while ROOT.name != "mia-tutorials" and ROOT.parent != ROOT:
    ROOT = ROOT.parent

DATA = ROOT / "data"
REPOS = ROOT.parent
ROOT

import sys
from collections import Counter

CLIMATE_REPO = REPOS / "mia-climate"
sys.path.insert(0, str(CLIMATE_REPO))

from lib.regions import validate_region, write_manifest
from lib.download_plan import build_download_manifest

REGION = DATA / "mia_climate" / "rbmn.geojson"
OUT = ROOT / "notebooks" / "mia_climate" / "outputs"
OUT.mkdir(parents=True, exist_ok=True)
REGION
```

```python
region_manifest = validate_region(
    region_id="rbmn",
    geometry_path=REGION,
    clip_policy="polygon",
    created_by="mia-tutorials/notebooks/mia_climate/01_validate_region_and_plan_downloads.ipynb",
)
write_manifest(OUT / "region_manifest.json", region_manifest)

region_manifest
```

```python
download_manifest = build_download_manifest(
    region_manifest=region_manifest,
    region_manifest_path=OUT / "region_manifest.json",
    start_year=2024,
    end_year=2024,
    created_by="mia-tutorials/tutorial",
)
write_manifest(OUT / "download_manifest.json", download_manifest)

counts = Counter(r["request_kind"] for r in download_manifest["requests"])
print(counts)
print("requires network:", download_manifest["requires_network"])
print("first request:")
download_manifest["requests"][0]
```
