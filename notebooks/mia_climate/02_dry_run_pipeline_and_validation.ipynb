# Dry-run Pipeline and Validation

This notebook shows how the script-based `mia-climate` workflow should be orchestrated from tutorials. It uses subprocess calls against the source repo so imports and relative paths behave like the real scripts.


```python
from pathlib import Path

ROOT = Path.cwd().resolve()
while ROOT.name != "mia-tutorials" and ROOT.parent != ROOT:
    ROOT = ROOT.parent

DATA = ROOT / "data"
REPOS = ROOT.parent
ROOT

import subprocess
import sys
from pathlib import Path

CLIMATE_REPO = REPOS / "mia-climate"
OUT = ROOT / "notebooks" / "mia_climate" / "outputs"
OUT.mkdir(parents=True, exist_ok=True)
PIPELINE_MANIFEST = OUT / "pipeline_manifest.json"
VALIDATION_REPORT = OUT / "validation_report.json"
```

## Source-repo packaging note

At the time this tutorial was scaffolded, `mia-climate` is script-based and has no visible top-level `pyproject.toml`. Running its scripts with `cwd=CLIMATE_REPO` is the least surprising pattern.


```python
RUN_DRY_PIPELINE = False

if RUN_DRY_PIPELINE:
    subprocess.run([
        sys.executable,
        "scripts/05_run_pipeline.py",
        "--mode", "dry-run",
        "--output", str(PIPELINE_MANIFEST),
        "--output-root", str(OUT),
    ], cwd=CLIMATE_REPO, check=True)
    print(PIPELINE_MANIFEST)
else:
    print("Skipped script execution. Set RUN_DRY_PIPELINE = True after checking script CLI arguments in the source README.")
```

```python
RUN_VALIDATION = False

if RUN_VALIDATION:
    subprocess.run([
        sys.executable,
        "scripts/06_validate_outputs.py",
        "--pipeline-manifest", str(PIPELINE_MANIFEST),
        "--output", str(VALIDATION_REPORT),
        "--output-root", str(OUT),
        "--mode", "dry-run",
    ], cwd=CLIMATE_REPO, check=True)
    print(VALIDATION_REPORT)
else:
    print("Skipped validation script execution.")
```
