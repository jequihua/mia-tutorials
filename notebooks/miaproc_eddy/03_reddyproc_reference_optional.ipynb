# Optional REddyProc Reference

This tutorial documents the optional R-backed REddyProc path. It is intentionally guarded because it requires R, `rpy2`, and a project-scoped REddyProc setup.


```python
from pathlib import Path

ROOT = Path.cwd().resolve()
while ROOT.name != "mia-tutorials" and ROOT.parent != ROOT:
    ROOT = ROOT.parent

DATA = ROOT / "data"
REPOS = ROOT.parent
ROOT

import os
from pathlib import Path

MIAPROC_REPO = REPOS / "miaproc"
eddy_dir = DATA / "miaproc_eddy"
MIAPROC_REPO, os.environ.get("R_HOME")
```

## Preflight first

The source example refuses to run unless the R runtime passes a project-scoped preflight. Run this cell only after installing the `miaproc[reddyproc]` extra and confirming the R setup.


```python
RUN_REDDYPROC_PREFLIGHT = False

if RUN_REDDYPROC_PREFLIGHT:
    from miaproc.eddy import (
        RRuntimePreflightPolicy,
        preflight_reddyproc_r_environment,
        render_r_preflight_report,
    )

    policy = RRuntimePreflightPolicy(repo_root=str(MIAPROC_REPO))
    result = preflight_reddyproc_r_environment(policy=policy)
    print(render_r_preflight_report(result))
else:
    print("Skipped. Set RUN_REDDYPROC_PREFLIGHT = True after R/rpy2/REddyProc are ready.")
```

## Reference run shape

Once preflight is approved, the reference path mirrors `miaproc/examples/03_r_backed_reddyproc_reference.py`: load stage 1, create a `ReddyProcConfig`, and call `postproc(engine="reddyproc-rpy2")`.


```python
RUN_REDDYPROC_BACKEND = False

if RUN_REDDYPROC_BACKEND:
    from miaproc.eddy import ReddyProcConfig, load_stage1, postproc

    stage1 = load_stage1(
        path_full_output=eddy_dir / "full_output",
        path_biomet=eddy_dir / "biomet",
        skip_full_output=0,
        skip_biomet=0,
        drop_rain_rows=False,
    )
    cfg = ReddyProcConfig(
        site_name="Marismas_Nacionales",
        latitude=22.25,
        longitude=-105.50,
        timezone_hour=-7.0,
        local_tz="America/Mazatlan",
    )
    out = postproc(stage1, engine="reddyproc-rpy2", reddyproc_config=cfg)
    print(out.attrs.get("miaproc_diagnostics"))
else:
    print("Skipped. Set RUN_REDDYPROC_BACKEND = True only after preflight is approved.")
```
